```python
import pandas as pd
file1_path = 'companies.txt'
companies_data = pd.read_csv(file1_path, sep='\t')
companies_data
```

```python
file2_path = 'rounds2.csv'
rounds2_data = pd.read_csv(file2_path, sep = ',')
rounds2_data
```

```python
unique_company = rounds2_data['company_permalink'].unique()
print(unique_company)
len(unique_company)
```

```python
unique_company2 = companies_data['permalink'].unique()
print(unique_company2)
len(unique_company2)
```

```python
print(companies_data['permalink'].str.upper())
print(rounds2_data['company_permalink'].str.upper())
```

```python
# Extract unique company names from both DataFrames
rounds2_companies = set(rounds2_data['company_permalink'].str.lower())
companies_companies = set(companies_data['permalink'].str.lower())

# Find companies in rounds2 that are not in companies.txt
companies_not_in_companies_txt = rounds2_companies - companies_companies

print("Companies in rounds2 that are not present in companies.txt:")
# print(companies_not_in_companies_txt)
rounds2_companies - companies_companies
```

```python
#to marge to dataframe we need similar looking common column
df1=rounds2_data.applymap(lambda x: x.lower() if isinstance(x, str) else x)

df2=companies_data.applymap(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
# type(rounds2_data)
# df1=rounds2_data.columns.str.lower()
# df2=pd.DataFrame(companies_data.columns.str.lower())
# print(df1)
# print(df2.head)
# Merge the two DataFrames using a left join on a common column (e.g., 'company_permalink' in rounds2 and 'permalink' in companies)
master_dataframe = pd.merge(df1, df2, left_on='company_permalink', right_on='permalink', how='left')
print(master_dataframe)
# Count the number of observations in master_dataframe
num_observations = master_dataframe.shape[0]
print("Number of observations in master_dataframe:", num_observations)
```

```python
# Drop the redundant column 'company_permalink' from master_dataframe
master_dataframe.drop('company_permalink', axis=1, inplace=True)
print(master_dataframe)
```

```python
# Calculate the percentage of missing values for each column
missing_percentages = (master_dataframe.isnull().sum() / len(master_dataframe)) * 100

#We create a DataFrame called missing_data to store the column names and their corresponding missing percentages.
# Create a DataFrame to display the missing percentages
missing_data = pd.DataFrame({'Column': master_dataframe.columns, 'MissingPercentage': missing_percentages})


# Display the columns with their missing percentages
print(missing_data)

#We sort the missing_info DataFrame in descending order of missing percentages to prioritize columns with higher missing values.
# Sort the columns by missing percentage in descending order
missing_data = missing_data.sort_values(by='MissingPercentage', ascending=False)

#We define a threshold (threshold) for missing values percentage. Columns with missing values exceeding this threshold will be dropped.
# # Define a threshold for missing values
threshold = 50

# # Create a list of columns to drop based on the threshold
columns_to_drop = missing_data[missing_data['MissingPercentage'] > threshold]['Column'].tolist()

# # Drop the unnecessary columns from master_dataframe
master_dataframe.drop(columns=columns_to_drop, inplace=True)


print('After dropping...')
print(master_dataframe)
```

```python
# Create a pivot table to compare mean and median values
pivot_table = master_dataframe.pivot_table(values=['raised_amount_usd'],
                                           index='funding_round_type',
                                           aggfunc={'raised_amount_usd': ['mean', 'median']})

# Reset the column names for the pivot table
pivot_table.columns = ['Mean', 'Median']

# Display the pivot table
print(pivot_table)
```

```python
# Filter master_dataframe based on 'raised_amount_usd'
filtered_master_dataframe = master_dataframe[(master_dataframe['raised_amount_usd'] >= 4000000) & (master_dataframe['raised_amount_usd'] <= 15000000)]
print(filtered_master_dataframe)
```

```python
# Filter the data for 'venture' fund type
venture_investments = filtered_master_dataframe[filtered_master_dataframe['funding_round_type'] == 'venture']

# Group the data by country and calculate total investment
country_investment = venture_investments.groupby('country_code')['raised_amount_usd'].sum()

# Sort the countries by total investment in descending order
top_countries = country_investment.sort_values(ascending=False)

# Select the top 9 countries
top_9_countries = top_countries.head(9)

# Print the top 9 countries and their total investments
print("Top 9 Countries in terms of Highest Investment in 'venture' Fund Type:")
print(top_9_countries)
```

```python
#We use the apply function on the 'category_list' column to apply a lambda function to each row.

#Inside the lambda function, we use the split('|') method to split the categories by the '|' delimiter, assuming that '|' separates different categories in the 'category_list'.

#We then take the first element of the split result as the main category using [0].

#We use isinstance(x, str) to check if the value in the 'category_list' column is a string to handle cases where it might be missing or not provided. If it's not a string (e.g., NaN values), we assign None to the 'main_category' column for those rows.

This code will create a new 'main_category' column in your DataFrame containing the extracted main sectors.
# Extract the main category from the 'category_list' column
companies_data['main_category'] = companies_data['category_list'].apply(lambda x: x.split('|')[0] if isinstance(x, str) else None)

# Display the updated DataFrame
print(companies_data)
```

```python
# Define a function to categorize 'raised_amount_usd' values
def categorize_amount(raised_amount_usd):
    if raised_amount_usd < 5000000:
        return 'TypeA'
    elif 5000000 <= raised_amount_usd < 7000000:
        return 'TypeB'
    else:
        return 'TypeC'

# Apply the categorization function to create the 'amount_category' column
filtered_master_dataframe['amount_category'] = filtered_master_dataframe['raised_amount_usd'].apply(categorize_amount)

# Display the DataFrame with the new 'amount_category' column
print(filtered_master_dataframe[['raised_amount_usd', 'amount_category']])
```
